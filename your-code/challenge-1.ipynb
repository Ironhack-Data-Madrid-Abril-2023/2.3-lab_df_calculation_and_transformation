#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
display(pokemon.head())

display(pokemon['Name']) #esto me printea unicamente la columna Name
```

```python
#funcion oriana


def cambio_mega(x):
    if "Mega " in x:
        x = 'Mega ' + x.split('Mega ')[1]
    return x


#Lo que he hecho a sido definir una función según x, que será el nombre de un pokemon, el la que:

#si aparece mega, x se convertira en un "nuevo mega" + una combinación de la palabra spliteada sin el mega antiguo
```

```python
#aquí lo que hago es aplicar la función a la columna Name de mi df pokemon
# ahora la columna pokemon.Name es igual a la columna actualizada
pokemon.Name = pokemon.Name.apply(cambio_mega)
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
def combine_types(row):
    if pd.notnull(row['Type 1']) and pd.notnull(row['Type 2']):
        return row['Type 1'] + '-' + row['Type 2']
    else:
        return row['Type 1']
    
#primero comprueba que se aplique a valores no nulos, y que escupa el t1-tipo 2
#si alguno es nulo, sólo pone tipo1

# ahora aplico la función a cada fila del dataframe para crear la nueva columna Combo Type
pokemon['Combo Type'] = pokemon.apply(combine_types, axis=1)

pokemon.head()
```
